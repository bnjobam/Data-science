```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("donut.csv",delimiter=" ", header=None)
```

```python
data.head()
```

```python
data.columns=["x1","x2", "y"]
```

```python
plt.scatter(data.x1,data.x2, c=data.y)
```

```python
def sigmoid(h):
    return 1/(1+np.exp(-h))

def accuracy(y, p_hat):
    return np.mean(y==np.round(p_hat))
```

```python
w=np.random.randn(3)
n=len(data.x1)
x=np.column_stack((np.ones(n),data.x1,data.x2))
X=x*x #h=Wx**2
p_hat=sigmoid(X.dot(w))
print('Accuracy: {:0.4f}'.format (accuracy(data.y,p_hat)))
```

```python
x1=np.linspace(-8,8,50)
x2=np.linspace(-8,8,50)
x1, x2=np.meshgrid(x1,x2)
plt.scatter(data.x1,data.x2, c=data.y)
F=w[0]+w[1]*x1**2+w[2]*x2**2
plt.contour(x1, x2,F, [0], colors='green')
```
